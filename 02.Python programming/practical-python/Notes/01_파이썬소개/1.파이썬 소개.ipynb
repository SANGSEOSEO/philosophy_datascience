## 1.2 첫번째 프로그램
어느 날 아침, 당신은 시카고의 시어스 타워(Sears tower) 근처를 거닐다가 보도에 1 달러 지폐를 한 장 올려뒀다. 그 후 매일 외출할 때마다 그 위에 지폐를 얹어 탑을 쌓으며, 높이는 매일 두 배로 불어난다. 돈으로 쌓은 탑의 높이가 시어스 타워의 높이와 같아지려면 시간이 얼마나 걸릴까?

```python
# sears.py
bill_thickness = 0.11 * 0.001 # 0.11은 지폐 1장의 두께, 0.001은 밀리미터를 미터로 변환
days = 1
number_of_bills = 1
towerHeight = 422
while bill_thickness * number_of_bills < towerHeight:
    print("현재까지 경과된 일은 {0}이고, 쌓인 지폐수는 {1:d}이며, 총 높이는 {2:.2f}".format(days, number_of_bills, number_of_bills * bill_thickness))
    days = days + 1
    number_of_bills = number_of_bills * 2
print("총 {0} 일 경과".format(days) )
print("총 지페수 : {0}".format(number_of_bills))
print("지폐의 높이 : {0}".format(bill_thickness * number_of_bills))
```

```python
# mortgage.py
principal = 500000.00 # 대출금액
rate = 0.05 # 이자율
payment = 2684.11 # 매달 이자
total_paid = 0.0

while principal > 0:
    principal = principal * (1 + rate / 12) - payment
    total_paid = total_paid + payment
print("Total paid", np.round(total_paid,1))
```

### 데이브의 주택담보 대출
- 추가 문제

데이브가 처음 12개월 동안 매달 1000 달러를 추가로 납입한다면 어떻게 될까?

이 추가 납입금 계산을 포함하도록 프로그램을 수정해 총 납입금액과 소요 월수를 프린트해보자.

프로그램을 실행했을 때 총 납입금이 929,965.62, 소요 월수는 342로 나와야 한다.

```python
principal = 500000.00 # 대출금액
rate = 0.05 # 이자율
payment = 2684.11 # 매달 이자
total_paid = 0.0
months = 0
additional_payment = 1000
while principal > 0:
    months = months + 1
    principal = principal * (1 + rate / 12) - payment
    total_paid = total_paid + payment
    
    if months <= 12:
        principal = principal - additional_payment
        total_paid = total_paid + additional_payment
        
print("Total paid", np.round(total_paid,2))
print("소요 월수 : ", months)
```

* 추가납입금 계산기 기능 추가

프로그램이 이 변숫값을 읽고 총 납입액을 계산하게 해 보자.

데이브가 대출 시작 5년 후부터 4년간 매달 1000 달러를 추가로 지불할 경우 총 납입액은 얼마인가?

```python
extra_payment_start_month = 61  # 대출개시일 5년시작
extra_payment_end_month = 108   # 대출 개시일부터 4년
extra_payment = 1000            # 1000달러 씩 추가 납부

principal = 500000.00 # 대출금액
rate = 0.05 # 이자율
payment = 2684.11 # 매달 이자
total_paid = 0.0
months = 0
while principal > 0:
    months = months + 1
    principal = principal * (1 + rate / 12) - payment
    total_paid = total_paid + payment
    
    if months >= extra_payment_start_month and months <= extra_payment_end_month:
        principal = principal - extra_payment
        total_paid = total_paid + extra_payment
print("데이브의 총 납입액 : {:.2f}".format(total_paid))        
```

#### 그외의 문자열 메소드 

```python
suffix = 'L'
s = "Hi, I am a data scientis who is specializing in ML and DL "
t = 'a'
old = "data"
new = "Data"
prefix = 'H'
slist = '-'

s.endswith(suffix)     # 문자열이 suffix로 끝나는지 확인
s.find(t)              # s에서 t가 처음 나타나는 곳
s.index(t)             # s에서 t가 처음 나타나는 곳
s.isalpha()            # 문자가 영문자인지 여부
s.isdigit()            # 문자가 숫자인지 여부
s.islower()            # 문자가 소문자인지 여부
s.isupper()            # 문자가 대문자인지 여부
s.join(slist)          # s를 구분자(delimiter)로 삼아 문자열의 리스트를 붙이기(join)
s.lower()              # 소문자로 변환
s.replace(old,new)     # 텍스트 교체
s.rfind(t)             # 문자열의 끝에서부터 t를 검색
s.rindex(t)            # 문자열의 끝에서부터 t를 검색
s.split(' ')       # 문자열을 분할해 부분 문자열의 리스트를 만듦
s.startswith(prefix)   # 문자열이 prefix로 시작하는지 확인
s.strip()              # 앞뒤의 공백을 제거
s.upper()              # 대문자로 변환
```

#### 4.문자열에 변수값을 포함하고 싶을때
* `mortgage.py`을 f'포맷 스트링으로 변경해봄.

```python
name = "IBM"
shares = 100
price = 91.1
f'{shares} shares of {name} at {price:.2f}'
```

```python
# mortgage.py
#포맷문자열을 이용해 변경
# 마지막달에 초과납부하지 않도록 로직 수정
extra_payment_start_month = 61  # 대출개시일 5년시작
extra_payment_end_month = 108   # 대출 개시일부터 4년
extra_payment = 1000            # 1000달러 씩 추가 납부

principal = 500000.00 # 대출금액
rate = 0.05 # 이자율
payment = 2684.11 # 매달 이자
total_paid = 0.0
months = 0
while principal > 0:
    months = months + 1
    principal = principal * (1 + rate / 12) - payment
    total_paid = total_paid + payment
    
    if months >= extra_payment_start_month and months <= extra_payment_end_month:
        principal = principal - extra_payment
        total_paid = total_paid  + extra_payment
    
    #print(months, total_paid, principal)
    
    # 매달 지불되는 금액과 현재 남은 원금비교해
    # 매달 지불해야 하는 금액이 크면 남은 원금을 직접결재하라고 하고 종료
    if extra_payment > principal:
        print(f"남은 금액이 ${principal:.2f} 입니다. 마지막 납부이므로 은행에 가서 직접 결재 필요")
        break;    
    
print(f"데이브의 총 납입액 : ${round(total_paid, 2)} 이며 , 총 납부한 월수는 {months:3d}")
```

```python
# csv로 읽어봄
import csv
target_dir = "../../Work/Data/"
def portpolio_cost(filename):
    total_cost = 0.0
    """
    주식매수시 총 가격 계산
    """
    with open(target_dir+filename, "rt") as f:
        headers = next(f)
        rows = csv.reader(f)
        for row in rows:
            try:
                shares = int(row[1])
                prices = float(row[2])
                total_cost += shares * prices
            except ValueError:
                print("Bad value Error")
        return total_cost
```

### 커스텀함수

```python
def sumcount(n):
    """
    1부터 n까지의 합을 반환
    """
    total = 0
    while n > 0:
        total += n
        n = n - 1
    return total
```
