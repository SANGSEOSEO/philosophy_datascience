```python
# pcost.py
import csv
target_dir = "../../Work/Data/"
def portpolio_cost(filename):
    total_cost = 0.0
    """
    ì£¼ì‹ë§¤ìˆ˜ì‹œ ì´ ê°€ê²© ê³„ì‚°
    """
    with open(target_dir+filename, "rt") as f:
        headers = next(f)
        rows = csv.reader(f)
        for row in rows:
            try:
                shares = int(row[1])
                prices = float(row[2])
                total_cost += shares * prices
            except ValueError:
                print("Bad value Error")
        return total_cost

import sys
if len(sys.argv) == 2:
    filename = sys.argv[1]
else:
    filename = input("Enter a filename:")
    
cost = portpolio_cost(filename)
print("Total Cost :", cost)
```
