# 오차 수정 : 경사하강법
* 반복적으로 기울기(a)를 변화시켜서 m의 값을 찾아내는 기법
* 기울기의 부호를 바꿔 이동시킬 때 적절한 거리를 찾지 못해 너무 멀리 이동시키면 a값이 한 점으로 모이지 않고(수렴하지 않음) 발산해버림.
따라서, 어느 만큼 이동시킬지를 신중히 결정해야 하는데, 이때 이동거리를 정해주는것이 바로 **학습률**이다.


### REPL
* 코딩으로 확인하는 경사하강법

```python
import numpy as np
```

```python
a , b = 0, 0
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]

# 학습률 
lr = 0.03
```

```python
x_data = np.array(x)
y_data = np.array(y)
```

```python
y_pred = a * x_data + b  # 오차함수인 y = ax + b를 정의
error = y_data - y_pred  # 실제값 - 예측값, 즉 오차를 구하는 식
```

```python
#평균제곱 오차를 a로 미분한 결과
a_diff = -(2 / len(x_data)) * sum(x_data * error)
b_diff = -(2 / len(x_data)) * sum(error)
```

```python
a = a - lr * a_diff  # 미분 결과에 학습율을 곱한 후 기존의 a 값을 갱신
b = b - lr * b_diff  # 미분 결과에 학습률을 곱한 후 기존의 b 값을 갱신
```

## 경사하강법 실습

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# 공부 시간 x, 성적 y의 리스트
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
def show_linear_reg(x, y):
  '''
  경사하강법
  '''
  # plot scatter
  plt.figure(figsize = (8, 5))
  plt.scatter(x, y)
  plt.show()

  x_data = np.array(x)
  y_data = np.array(y)

  # slope - a, intersect - b 
  a , b = 0, 0

  # learning_rate 
  lr = 0.03

  # epochs
  epochs = 2001

  # Gradient Descent
  for i in range(epochs):
    y_pred = a * x_data + b
    error = y_data - y_pred

    a_diff = -(2 / len(x_data)) * sum(x_data * error)
    b_diff = -(2 / len(x_data)) * sum(error)

    a = a - lr * a_diff  # 미분 결과에 학습율을 곱한 후 기존의 a 값을 갱신
    b = b - lr * b_diff  # 미분 결과에 학습률을 곱한 후 기존의 b 값을 갱신

    if i % 100 == 0:
      print("epochs= %.f, slopes = %.04f, intercept = %.04f" %(i, a, b))
  
  # 앞서 구한 기울기와 절편을 이용해 그래프 다시 그리기
  y_pred = a * x_data + b
  plt.scatter(x, y)
  plt.plot([min(x_data), max(x_data), min(y_pred), max(y_pred)])
  plt.show()
```

```python
# 수행
show_linear_reg(x, y)
```
