## 결측치 처리 및 전처리,정규화
* 붓꽃 데이터셋 이용

```python
data = pd.read_csv("../04.Pandas/data/iris.csv")
data.head()
```

```python
data.describe()
```

### 널값 갯수 확인

```python
data.isnull().sum()
```

### 각 피처의 평균값 확인

```python
sepal_length_mean = data["sepal.length"].mean()
sepal_width_mean = data["sepal.width"].mean()
petal_length_mean = data["petal.length"].mean()
petal_width_mean = data["petal.width"].mean()
```

### 널인 값들은 평균값으로 대체

```python
data["sepal.length"].fillna(sepal_length_mean, inplace = True)
data["sepal.width"].fillna(sepal_width_mean, inplace = True)
data["petal.length"].fillna(petal_length_mean, inplace = True)
data["petal.width"].fillna(petal_width_mean, inplace = True)
```

### 정규화 함수 생성

```python
def normalizer(columnname):
    max = data[columnname].max()
    min = data[columnname].min()
    
    return (data[columnname]- min) / (max - min)

data['sepal.length'] = normalizer('sepal.length')
data['sepal.width'] = normalizer('sepal.width')
data['petal.width'] = normalizer('petal.width')
print(data)
```

### 구간 나누기 
* np.histogram사용(구간을 나누는 함수)
* 데이터의 범위를 알아내서 지정된 구간을 나누고 구간의 경계에 대한 값을 가져옴
* 구간을 나누고자 하는 필드와 구간의 개수를 전달하면 구간의 경계 , 구간의데이터 개수를 반환

```python
help(np.histogram)
```

```python
count, bin_dividers = np.histogram(data['petal.length'], bins = 3)
```

```python
count, bin_dividers
```

* cut 함수 , x 인자에 나누고자 하는 구간의 필드를 정해줌
* bins 인자에 np.histogram 이 생성한 구간 정보를 전달
* labels 에 각 구간에 붙일 라벨명을 지정
* include_lowest =True 는 경계값을 포함하라는 의미

```python
bin_names = ["A", "B", "C"]
data['petal_grade'] = pd.cut(data['petal.length'], bins = bin_dividers, labels = bin_names, include_lowest = True)
print(data)


Y_class = np.array(data['petal.length']).reshape(-1, 1) # 2차원으로 바꿈

from sklearn.preprocessing import OneHotEncoder

# 데이터 필드가 1차원 상태로는 One-Hot Encoding이 진행되지 않음
# reshape함수는 1차원 Series타입을 2차원의 ndarray타입으로 변환
enc = OneHotEncoder()

enc.fit(Y_class)
Y_class_one_hot = enc.transform(Y_class).toarray()

# argmax()함수 - 온핫 인코딩 복원시 사용, 전달받은 인자 중 가장 큰 값을 갖는 인자를 반환하는 함수.
Y_class_recovery = np.argmax(Y_class_one_hot, axis = 0).reshape(-1, 1)

print(Y_class_one_hot)
```
