### 1.1 Null Data Check

```python
def null_data_check(df):
    """
    Null Data Check
    """
    for col in df.columns:
        msg = 'column : {0:>10} \t Percent of null value : {1:.2f}'.format(col, 100 * df[col].isnull().sum() / df.shape[0])
        display(msg)
```

```python
null_data_check(df_train)
```

```python
null_data_check(df_test)
```

```python
#널 체크 라이브러리로서 널데이터를 시각화해서 보여주는 것이 가능
import missingno as msno
def visualize_null_data(df, opt='matrix'):
    """
    해당 데이터셋의 널값에 대한 시각화
    """
    if opt=='matrix':
        # 매트릭스로 표시
        msno.matrix(df.iloc[:, :], figsize=(8, 4), color=(0.8, 0.5, 0.3))
    else:
        msno.bar(df.iloc[:, :], figsize=(8, 4), color=(0.8, 0.5, 0.3))
```

### 1.2 Target Label 확인
* 타겟값의 분포도를 확인하여 편향된 분포도를 갖는지 확인이 필요
* 이진분류(binary classification)의 경우 1과 0의 분포도에 따라 평가방법이 달라질수 있음

```python
def visualization_target(df):
    """
    Target에 대한 데이터분포도 시각화
    """
    f, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
    survived = df['Survived'].value_counts()
    survived.plot.pie(explode = [0, 0.1], autopct='%1.1f%%', ax = ax[0], shadow=True)
    ax[0].set_title("Pie Plot - Survived")
    ax[0].set_ylabel('')
    sns.countplot(x='Survived', data = df, ax = ax[1], palette='Paired')
    ax[1].set_title("Count Plot - Survived")
    plt.tight_layout()
    plt.show()
    
# 타겟값에 대한 시각화
visualization_target(df_train)
```

### 2.7 Sibsp, Parch
* 두 피처를 합친 새로운 피처 생성할것(FamilySize)

```python
def feature_engineering(df_train, df_test, logT=False):
    """
    피처 엔지니어링
    1.FamilySize컬럼 추가 - 가족수
    2.Fare feature의 널값 처리 
    3.편향되어 있는 Fare피처에 대해서 로그변환(np.log1p)
    """
    df_train.drop('FamilySize', axis=1, inplace=True, errors='ignore')
    df_test.drop('FamilySize', axis=1, inplace=True, errors='ignore')   
    
    df_train['FamilySize'] = df_train['SibSp'] + df_train['Parch'] + 1
    df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch'] + 1  
    
    display(df_train.head(n=2))
    display("가장 많은 가족수 : \t", df_train['FamilySize'].max())
    display("가장 적은 가족수 : \t", df_train['FamilySize'].min())
    
    # 널인 Fare값을 평균값으로 대체
    df_train['Fare'] = df_train['Fare'].fillna(value=df_train['Fare'].mean(), axis=0)
    df_test['Fare'] = df_train['Fare'].fillna(value=df_test['Fare'].mean(), axis=0)
    
    # 편향되어있는 Fare 피처를 로그변환 처리 
    if logT:
        df_train['Fare'] = df_train['Fare'].apply(lambda x: np.log1p(x))
        df_test['Fare'] = df_test['Fare'].apply(lambda x: np.log1p(x))
    
    return df_train, df_test
```

```python
df_train, df_test = feature_engineering(df_train, df_test)
```

* 가족수와 생존률의 관계

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(40, 10))
sns.countplot(x='FamilySize', data = df_train, ax=ax[0])
ax[0].set_title("(1) No of Passengers boarded", y = 1.02)

sns.countplot(x='FamilySize', hue='Survived', data=df_train, ax=ax[1])
ax[1].set_title("(2) Survived countplot depending on the family size")

familysize_survived_mean = df_train[['FamilySize', 'Survived']].groupby(by='FamilySize', as_index=True, squeeze=False).mean()
familysize_survived_mean.sort_values(by='Survived', ascending=False).plot.bar(ax=ax[2])
ax[2].set_title("(3) Survived rate depending on Family size")
plt.subplots_adjust(wspace = 0.2, hspace=0.5)

plt.show()
```

* 가족 수가 많다고 해서 생존률이 높은것은 아니지만, 2명~4명사이의 생존률이 제일 높다고 할 수 있다.

* 요금대로 알아보는 생존률

```python
def visualization_distplot(df):
    """
    hist와 kdeplot을 합친 그래프 
    """
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    
    fare_data = df['Fare']
    g = sns.distplot(a=fare_data, color='r', label="skewness: {:.2f}".format(fare_data.skew()), ax=ax)
    g.legend(loc='best')
    plt.grid(True)
    plt.show()
```

```python
visualization_distplot(df_train)
```

* 요금이 0 ~ 50사이에 집중되어 있으며 편향된 분포를 보인다. 이진분류인 경우에는 로그변환을 통해 정규화분포를 만들고 예측모델을 생성하는것이 필요해 보임
