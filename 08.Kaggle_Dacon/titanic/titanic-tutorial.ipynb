```python
def null_data_check(df):
    """
    NaN data check
    """
    for col in df.columns:
        msg = "column : {0:>10} \t Percent of null value : {1:.2f}".format(col, 100 * train[col].isnull().sum() / train.shape[0])
        display(msg)
        
# call
null_data_check(train)
```

```python
def visualization_target(df):
    """
    Target에 대한 데이터분포도 시각화
    """
    f, ax = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
    survived = df['Survived'].value_counts()
    survived.plot.pie(explode = [0, 0.1], autopct='%1.1f%%', ax = ax[0], shadow=True)
    ax[0].set_title("Pie Plot - Survived")
    ax[0].set_ylabel('')
    sns.countplot(x='Survived', data = df, ax = ax[1], palette='Paired')
    ax[1].set_title("Count Plot - Survived")
    plt.tight_layout()
    plt.show()
    
# 타겟값에 대한 시각화
visualization_target(train)
```

### 2.4 Age

* 연령대

```python
def show_age_information(df):
    """
    1.최연장자와 최연소 나이 및 평균승선자의 연령
    2.연령별 생존률 비교
    3.객실등급별 연령대 비교
    4.나이를 증가시키면서 생존률 비교 시각화
    """
    display("최연장자 나이 : {}".format(df["Age"].max()))
    print("\n", "최연소 나이")
    display(df["Age"].min())
    print("\n", "평균 승선자 연령 {}".format(train["Age"].mean()))
    
    # 2.연령별 생존률 비교
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 5))
    sns.kdeplot(df[df["Survived"]== 1]["Age"], ax = ax)
    sns.kdeplot(df[df["Survived"]== 0]["Age"], ax = ax)
    plt.legend(["Survived == 1", "Survived == 0"])
    
    # 3.승객의 객실등급별 연령대 분포
    cond1 = df["Pclass"]==1
    cond2 = df["Pclass"]==2
    cond3 = df["Pclass"]==3
    
    plt.figure(figsize=(12, 5))
    df[cond1]["Age"].plot(kind='kde')
    df[cond2]["Age"].plot(kind='kde')
    df[cond3]["Age"].plot(kind='kde')
    plt.xlabel("Age")
    
    plt.title("Age Distribution within Pclass")
    plt.legend(["1st class", "2nd class", "3rd class"])    
    plt.show()
    
    # 4.연령별 누적 생존률 비교
    y_position = 1.02
    cumulative_survivor_ratio = []
    
    for i in range(1, 80):
        cond1 = df["Age"] < i
        survivor_ratio = df[cond1]["Survived"].sum() / len(df[cond1]["Survived"])
        cumulative_survivor_ratio.append(survivor_ratio)
        
    plt.figure(figsize=(12, 5))
    plt.plot(cumulative_survivor_ratio)
    plt.xlabel("Range of Age from 0 to 80")
    plt.ylabel("Cumulative Survivor Ratio")
    plt.title("Cumulative Survivor Ratio as getting old", y = y_position)
    plt.show()
```

```python
show_age_information(train)
```

### 2.7 Sibsp, Parch
* 두 피처를 합친 새로운 피처 생성할것(FamilySize)
* 가장 가족수가 많은 것과 가장 적은 것 보여줘

```python
# 피처 추가 
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
test["FamilySize"] = test["SibSp"] + test["Parch"] + 1
```

```python
train.head(n=3)
```

```python
def feature_engineering(train, test, logT=False):
    """
    피처 엔지니어링
    1.FamilySize컬럼 추가 - 가족수
    2.Fare feature의 널값 처리 
    3.편향되어 있는 Fare피처에 대해서 로그변환(np.log1p)
    """
    
    # 피처 추가로 인해 불필요해진 컬럼 삭제 
    print("\n 삭제전")
    display(train.head(n= 3))
    train = train.drop(["SibSp", "Parch"], axis=1, inplace=False, errors="ignore")
    test = test.drop(["SibSp", "Parch"], axis=1, inplace=False, errors="ignore")
    
    print("\n 삭제후 ")
    display(train.head(n=3))
    
    display("가장 많은 가족수 {0} 와 가장 적은 가족수 {1}".format(train["FamilySize"].max(), train["FamilySize"].min()))
    
    # 결측치 처리 
    train["Fare"] = train["Fare"].fillna(value = train["Fare"].mean(), axis=0)
    test["Fare"] = test["Fare"].fillna(value = test["Fare"].mean(), axis=0)
    
    # 편향되어 있는 Fare값 로그변환 
    if logT:
        train["Fare"] = train["Fare"].apply(lambda x: np.log1p(x))
        test["Fare"] = test["Fare"].apply(lambda x: np.log1p(x))
        
    return train, test

# 수행 
train, test = feature_engineering(train, test)
```

* 가족수와 생존률의 관계
* [squeeze옵션 관련](https://www.inflearn.com/questions/160111)

```python
y_position = 1.02

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(40, 10))
sns.countplot(x="FamilySize", data=train, ax= ax[0])
ax[0].set_title("1.Nbr of Passengers boarde on Ship", y=y_position)

sns.countplot(x="FamilySize", hue="Survived", data=train, ax = ax[1])
ax[1].set_title("2.Survivor countPlot depending on Family size", y=y_position)

familysize_survived_mean = train[["FamilySize", "Survived"]].groupby(["FamilySize"], as_index=True, squeeze=False).mean()
display(familysize_survived_mean)

familysize_survived_mean.sort_values(by="Survived", ascending=False).plot.bar(ax= ax[2])
ax[2].set_title("3.Survivor Ratio depending on Family Size.")
plt.subplots_adjust(wspace= 0.2, hspace=0.5)
plt.show()
```

* 가족 수가 많다고 해서 생존률이 높은것은 아니지만, 2명~4명사이의 생존률이 제일 높다고 할 수 있다.
* 11명의 대가족인 경우엔, 비극이지만 비상상황시에 이동성에 문제가 있을 수도 있음에 유의미 의미가 있음.
* 1명인 경우의 생존률이 의외로 높지 않음에 주목(사망자 수 - 350명이 훨씬 상회함)

```python
def visualization_distplot(df):
    """
    hist와 kdeplot을 합친 그래프 
    """
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    
    fare_data = df['Fare']
    g = sns.distplot(a=fare_data, color='r', label="skewness: {:.2f}".format(fare_data.skew()), ax=ax)
    g.legend(loc='best')
    plt.grid(True)
    plt.show()
```
