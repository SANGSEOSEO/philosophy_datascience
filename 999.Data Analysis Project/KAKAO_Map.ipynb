```python
# 카카오 검색 API
import requests

searching = 'Espania'
#40.46, -3.75
url = "https://dapi.kakao.com/v2/local/search/keyword.json?&query={}&x=40.46&y=-3.75".format(searching)
headers = {"Authorization" : "KakaoAK 40c2c3974984e33103edf8eeed97380a"} # values사이에는 반드시 한칸 띄우어 하야 함.


places = requests.get(url, headers = headers).json()['documents']
places

#서귀포잠수함	126.558616	33.239303	Seogwipo	
```

### 함수로 생성

```python
def find_places(searching):
    url = "https://dapi.kakao.com/v2/local/search/keyword.json?&query={}".format(searching)
    headers = {"Authorization" : "KakaoAK 40c2c3974984e33103edf8eeed97380a"} # values사이에는 반드시 한칸 띄우어 하야 함.

    places = requests.get(url, headers = headers).json()['documents']
    
    #print(type(places))
    place = places[0]
    name  = place['place_name']
    x = place['x']
    y = place['y']
    data = [name, x, y , searching]
    return data
```

```python
# 제주공항의 위치 정보를 찾아 위경도를 리턴
find_places('제주공항')
```

```python
from tqdm import tqdm_notebook
```

```python
import time
locations_inform = []


for location in tqdm_notebook(locations):
    try:
        data = find_places(location)
        locations_inform.append(data)
        time.sleep(0.5)
    except:
        pass
locations_inform
```

## folium을 이용한 지도 시각화 - 1개별 표시

```python
location_data.head(n = 1)

#location_data['name_official'][0] # 이렇게 하면 에러남.
location_data.index
```

```python
import folium
```

```python
Mt_Hanla = [33.362500, 126.533694]
map_jeju = folium.Map(location = Mt_Hanla, zoom_start = 11)

for i in range(len(location_data)):
        name = location_data['name_official'][i]  # 공식명칭
        count = location_data['place'][i]  # 게시글 개수
        size = int(count) * 2  
        long = float(location_data['위도'][i])
        lat = float(location_data['경도'][i])
        folium.CircleMarker((long, lat),
                           radius = size, color = 'red', popup = name).add_to(map_jeju)
    
map_jeju
```

```python
from folium.plugins import MarkerCluster
locations = []
names = []


for i in range(len(location_data)):
    data = location_data.iloc[i]
    locations.append((float(data['위도']), float(data['경도']))) 
    names.append(data['name_official'])
    
Mt_Hanla = [33.362500, 126.533694]
map_jeju2 = folium.Map(location = Mt_Hanla, zoom_start = 11)

marker_cluster = MarkerCluster(locations = locations, 
                              popups = names,
                              name = 'Jeju',
                              overlay = True,
                              control = True)
marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)
map_jeju2
```

```python
map_jeju.save("../data/3_jeju.html")
```
