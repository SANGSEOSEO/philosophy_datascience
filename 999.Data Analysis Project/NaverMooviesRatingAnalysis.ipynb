```python
# 주말, 공유일까지 다 포함하는 모든 날.
date = pd.date_range('2017-05-01', periods = 100, freq = 'D')
date
```

```python
import urllib
from tqdm import tqdm_notebook
movie_date = []
movie_name = []
movie_point = []

for today in tqdm_notebook(date):
    html = "http://movie.naver.com/" +\
                                "movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date = urllib.parse.quote(today.strftime("%Y%m%d"))))
    
    soup = BeautifulSoup(response,"html.parser")
    end = len(soup.find_all('td', 'point'))
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)])
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date': movie_date, 'name': movie_name,'point':movie_point})
movie.head()
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
movie.info()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum)
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
```

```python
tmp = movie.query('name == ["노무현입니다"]')
tmp
```

```python
import matplotlib.pyplot as plt
#matplotlib inline

plt.figure(figsize = (12, 8))
plt.plot(tmp['date'], tmp['point'])
plt.legend(loc = 'best')
plt.grid()
plt.show()
```

```python
import pandas as pd
date = pd.date_range(start = '2020-02-01', periods = 100)
date
```

```python
import urllib
from tqdm import tqdm_notebook
movie_date = []
movie_name = []
movie_point = []

# urllib.parse.quote - 호출 url에서 url request로 파라미터 전달

for today in tqdm_notebook(date):
    html = "http://movie.naver.com/" +\
                                "movie/sdb/rank/rmovie.nhn?sel=cur&date={date}"
    response = urlopen(html.format(date = urllib.parse.quote(today.strftime("%Y%m%d"))))
    
    soup = BeautifulSoup(response,"html.parser")
    end = len(soup.find_all('td', 'point'))
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div', 'tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td', 'point')[n].string for n in range(0, end)]) 
```

```python
len(movie_date), len(movie_name), len(movie_point)
```

```python
movie = pd.DataFrame({'date': movie_date, 'name': movie_name,'point':movie_point})
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
```

```python
movie['name'].unique()
```

```python
hotList = ["퐁네프의 연인들", "포레스트 검프", "설국열차", "인셉션", "대통령의 7시간", "기생충", "남산의 부장들"]
movieHotList = movie[movie['name'].isin(hotList)]
movieHotList
```

```python
movie_pivot = pd.pivot_table(movieHotList, index=["date"], columns=['name'], 
                                         values=['point'])
movie_pivot.head()
```

```python
movie_pivot.columns = movie_pivot.columns.droplevel()
```

```python
movie_pivot.tail()
```

```python
plt.figure(figsize = (12, 5))
movie_pivot.plot()
plt.title("2020년 2월 이후 개봉 영화 평점 추이")
plt.xlabel("일자")
plt.ylabel("영화 평점")
plt.grid()
plt.show()
```
