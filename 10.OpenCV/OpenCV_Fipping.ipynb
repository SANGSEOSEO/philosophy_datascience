### 패키지 임포트 

```
# import the necessary packages
import matplotlib.pyplot as plt
import argparse
import cv2
import imutils
```

### 코랩 및 주피터 노트북에서 이미지 보기

```
def plt_imshow(title, image):
  """
  이미지를 BGR에서 RGB로 변환하고 해당 이미지를 리턴
  """
  image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
  plt.title(title)
  plt.imshow(image)
  plt.grid(False);plt.show()
```

### OpenCV를 이용한 이미지 뒤집기 

```
help(cv2.flip)
```

```
# # 파이썬으로 만들어진 스크립트를 콘솔에서 수행(파라미터 파싱)
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", type=str, default="opencv_logo.png",
# 	help="path to the input image")
# args = vars(ap.parse_args())

# 주피터 혹은 코랩에서 수행시 윗부분 주석 처리 
#args = {
#	"image": "opencv_logo.png"
#}
args = {
	"image": "IMG_4918.jpg"
}
```

```
# 이미지 로드
image = cv2.imread(args['image'])
plt_imshow("Original Image", image)

# 수평으로 이미지 뒤집기
print("[INFO] flipping image horizontally...")
flipped = cv2.flip(image, 1)  # `1`은 수평
plt_imshow("Flipped horizantly", flipped)
```

```
# 이미지 수직으로 뒤집기
flipped = cv2.flip(image, 0)  # `0`은 수직
print("[INFO] flipping image vertically...")
plt_imshow("Flipped Vertically Image", flipped)
```

```
# flip the image along both axes
flipped = cv2.flip(image, -1)  # 이미지를 수평, 수직으로 뒤집기
print("[INFO] flipping image horizontally and vertically...")
plt_imshow("vertically and horizontally style", flipped)
```
