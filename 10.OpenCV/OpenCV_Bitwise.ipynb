## 코드 개시 및 수행

### 패키지 임포트 

```
# import the necessary packages
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

### 코랩 및 쥬피퍼에서 도형 그리기

```
def plt_imshow(title, image):
  """
  이미지프레임을 BGR에서 RGB로 변환하고 해당 이미지를 보여줌
  """
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  plt.imshow(image)
  plt.title(title)
  plt.grid(True)
  plt.show()
```

### OpenCV에서 AND, NOT, OR, XOR 비트 연산

```
def show_figure(initPos):
  # 사각형 그리기
  rectangle = np.zeros(initPos, dtype='uint8')
  cv2.rectangle(rectangle, (25, 25), (275, 275), 255, -1) # 이미지, pt1, pt2, color, 직각_색깔채움
  plt_imshow("Rectangle", rectangle)

  # 원 그리기
  circle = np.zeros(initPos, dtype='uint8')
  cv2.circle(circle, (150, 150), 150, 255, -1)
  plt_imshow("Circle", circle)
```

```
show_figure((300, 300))
```

### 정리
* 두 이미지는 두 개의 픽셀강도 값만 가지고있음.0 (검정) 또는 픽셀이 0보다 큼(흰색) - 픽셀값이 두개인 값만 가지는 이미지를 바이너리 이미지하고 한다.
* 비트 연산 : 거실의 on/off 스위치의 기능을 기억하자.

```
# AND 비트연산은 두 입력값이 모두 "ON"일때는 참, 그렇지 않고 두 값이 서로 다르다면 `거짓`
bitwise_and_img = cv2.bitwise_and(rectangle, circle)
plt_imshow("Bitwise And Image", bitwise_and_img)
```

```
# OR 비트 연산 - 두 개의 입력값 중에 하나의 입력값만 참이더라도 결과값은 참
bitwise_or = cv2.bitwise_or(rectangle, circle)
plt_imshow("Bitwise OR Image", bitwise_or)
```

```
# XOR 연산은 두 픽셀 중 하나가 0보다 크지만 두 픽셀이 모두 0보다 클 수 없는 경우에만 참
# XOR 연산은 둘 중 하나만 참인 경우만 True이고 둘 다 참이면 False이다.
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
plt_imshow("bitwise_xor", bitwise_xor)
```

```
# `NOT`연산은 픽셀의 값을 기본적으로 뒤집습니다.
# 즉, 0보다 큰 픽셀은 0으로 바뀌고 , 0인 픽셀은 1로 바뀜
bitwise_not = cv2.bitwise_not(rectangle, circle)
plt_imshow("Bitwise Not ", bitwise_not)

# 이전 도형 
initPos = (300, 300)
show_figure(initPos)
```

### 정리
* 비트연산(AND , NOT, OR, XOR)은 알파블렌딩 혹은 매스킹 작업에 필요하므로 익혀두기 바랍니다.
