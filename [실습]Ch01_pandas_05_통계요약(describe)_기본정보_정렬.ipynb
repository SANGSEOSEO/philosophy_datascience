## 7. 정렬하기 

### 7-1. 오름차순 정렬 (default)

```
# 오름차순 정렬(default), ascending=True가 기본값
# 인덱스로 정렬
# inflacd=True는 원본을 정렬
df.sort_index()
```

### 7-2. 내림차순 index 정렬

```
# 내림차순 index정렬
df.sort_index(ascending=False)
```

## 8. column 별로 정렬

### 8-1. 오름차순 정렬

```
# 컬럼별로 오름차순 정렬
# 신장과 인지도 관련있는가?
df.sort_values(by='키')
```

### 8-2. 내림차순 정렬

```
# 코드를 입력해 주세요
df.sort_values(by='키', ascending=False)
```

### [BONUS] 복수 정렬

```
# 복수의 컬럼으로 정렬하고 기본 정렬 옵션이 올림차순
# ascending=True
df.sort_values(by=['키', 'brandNameValue'])
```

```
# 복수컬럼으로 정렬하고 내림차순으로 정렬
df.sort_values(by=['키', 'brandNameValue'], ascending=False)
```
