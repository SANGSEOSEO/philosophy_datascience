## 1. 결측값을 채워주는 fillna

**fillna()**: na 값에 대하여 fill해주는 함수입니다.
 - NaN인 값에 대해서 해당 값을 채워주어야 할 필요(전처리)

```
df.info()
```

키에 2개의 데이터가 누락, 그룹에 1개의 데이터가 누락된 것을 확인할 수 있습니다.

먼저, 키가 없는 사람은 존재할 수 없기 떄문에, 데이터가 잘 못 되어 있다는 것을 표기 하기 위하여, **누락된 데이터를 -1로 채워** 보도록 하겠습니다.

###1.1 NaN인것을 loc를 이용해 전처리 

```
df['키']
```

```
# isna() 인것을 loc를 이용해 데이터 전처리 
df.loc[df['키'].isna(), '키'] = -1
```

```
# 키가 NaN인것이 있는지 다시 확인
df[df['키'].isna()]
```

###1.2 fillna()를 이용해 NaN인 값은 특정값으로 처리 

```
# 바로 데이터프레임에 반영됨(inplace=True)
df['키'].fillna(-1, inplace=True)

# 혹은 다음과 같이 하면 바로 datafame에 반영됨
df['키'] = df['키'].fillna(-1)
```

```
# -1로 바뀐것이 있는지 찾아보자.
# 있으니 정상적으로 바뀌었어.
df.loc[df['키'] == -1]
```

NaN 값이 -1로 바뀐 것을 확인할 수 있습니다.

하지만, 키의 NaN 값을 채워준다음 유지시키려면 **inplace=True** 옵션을 주거나, fillna로 **채워 준 값을 다시 대입**해 주어야 합니다.

####방법1

```
df2 = df.copy()
```

```
df2.head()
```

```
df2['키'].fillna(-1,inplace=True)
```

```
df2.head()
```

```
df2['키']
```

이젠 유지가 됩니다. 하지만, 저는 **fillna로 채워준 값을 다시 대입 해 주는 방식을 추천**드립니다.

#### 방법2
 - fillna()채워준 값을 다시 대입하는 방식

```
df2 = df.copy()
```

```
df2.info()
```

```
df2['키']
```

```
df2['키'] = df2['키'].fillna(-1)
```

```
df2['키']
```

#### 방법3
 - 평균값 혹은 다른 통계치로 채워넣기

```
df2 = df.copy()
```

```
df2.head()
```

```
df2['키']
```

```
# 평균값.
height = df2['키'].mean()

# 최빈값
height = df2['키'].mode()
height
```

## 2. 빈 값(NaN)이 있는 행을 제거

1번 STEP에서 fillna 메소드를 활용하여 값을 채워 주었습니다. 이번에는 빈값이 있는 행을 제거하는 방법을 알아보도록 하겠습니다.

```
df.info()
```

```
df.head()
```

```
df2 = df.copy()
```

dropna()는 몇가지 옵션을 추가할 수 있습니다.

### 2-1. axis (열/ 행을 드랍)

axis=0은 행을 드랍합니다.

```
# 해당 행(row)가 삭제되고 df2 데이터 프레임에 바로 적용됨.
# how=`'any'` 옵션은 NaN이 하나라도 있으면 해당 로우는 삭제
df2.dropna(axis=0, how='any', inplace=True)
```

```
df2
```

axis=1은 열을 드랍합니다.

```
df2 = df.copy()
```

```
# 키 정보에 NaN값이 2개 있고, 그룹 정보에 NaN값이 하나 있다.
# 특정 열에 NaN값이 있으면 해당 열을 삭제
df2.dropna(axis=1, how='any', inplace=True)
```

```
# `키` 컬럼도 사라지고 `그룹` 컬럼도 사라졌다.
df2
```

### 2-2. how옵션 - 'any': 한개라도 있는 경우 드랍, 'all'은 모두 NaN인 경우 드랍
  - inplace옵션 : True - 원본에 적용, False - 기본값
  - how옵션 : default값은 any

```
df2 = df.copy()
```

```
df2.head(n=10)
```

```
# 모든 행(로우)가 NaN이어야 삭제됨
# 일부여야 삭제가 안됨
# df2.dropna(axis=0, how='all')  # 그 행에 전부 NaN이어야 삭제됨.
df2.dropna(axis=0, how='any')  # 행에 하나라도 NaN이 있으면 삭제
```

```
df2.dropna(axis=1, how='any')
```

```
df.dropna(axis=0, how='any')
```

```
df.dropna(axis=0, how='all')
```

## 3. 중복된 값 제거 (drop_duplicates)

```
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```
df.head()
```

### 3-1. column의 중복값 제거
  - subset : 중복을 제거하고자 하는 컬럼 sequence
  - keep 
      * first - 중복건준에 처음것을 제거
      * last  - 나중에 나온 것을 제거 
      * False - 모두 제거 

```
df['키']
```

```
# keep옵션 : defaul는 first로 먼저 나온값은 보존하고 나중에 나온값의 해당 로우는 삭제처리
#            last값은 먼저 나온 값의 로우 혹은 행은 삭제처리하고 나중에 나온 로우 혹은 컬럼은 보존
# keep옵션의 값을 아무것도 안주면 기본적으로 first이다.
df['키'].drop_duplicates()
```

```
df['키']
```

```
# 키를 비교했을때 중복되는 값을 처음에 나온것은 보존하고 나중에 나온것은 삭제
df['키'].drop_duplicates(keep='first',inplace=False)
```

```
# 보존하지 않고 무조건 중복 데이터는 삭제
df['키'].drop_duplicates(keep='last', inplace=False)
```

keep 옵션으로 유지하고 싶은 데이터를 선택할 수 있습니다. keep: 'first' / 'last'

```
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```
df['키']
```

```
df['키'].drop_duplicates(keep='last')
```

```
# 중복되는 것을 삭제하고 원본 데이터에 반영하기 
df['키'].drop_duplicates(keep='last', inplace=True)
```

```
df
```

### 3-2. 행 전체 제거

```
# 그룹이라는 행전체를 제거하는 기준에 `그룹`이라는 컬럼에 중복되는 정보가 있으면 
# 전체 행을 삭제한다.
df.drop_duplicates('그룹')
```

```
df.drop_duplicates('그룹', keep='first')
```

```
# `그룹`, `소속사` 값이 중복되면 마지막것만 남겨두고 행(로우)를 삭제
df.drop_duplicates(subset=['그룹','소속사'], keep='last')
```

```
df.drop_duplicates(subset='그룹', keep='first')
```

## 4. Drop - column/row 제거하기

```
df = pd.read_csv('https://bit.ly/ds-korean-idol')
```

```
df.head()
```

### 4-1. column 제거하기

drop()을 활용하여 column을 제거할 수 있습니다. column을 제거할 때는 axis=1 옵션을 줍니다.

```
# `그룹` 컬럼을 삭제
# `axis = 1` 은 열(컬럼)을 의미
df
df.drop(axis=1, columns=['생년월일', '키', '혈액형'])
```

복수의 column를 제거하고자 할 때는 list로 지정합니다.

```
df.drop(axis=1, columns=['그룹', '소속사'])
```

### 4-2. row 제거하기

drop()을 활용하여 row를 제거할 수 있습니다. row를 제거할 때는 제거하고자하는 index와 axis=0 옵션을 줍니다.

```
# 행(row) 제거하기 
df.drop(axis=0, index=[0, 1])
```

복수의 row를 제거하고자 할 때는 list로 지정합니다.

```
df.drop(axis=0, index=10)  
```
